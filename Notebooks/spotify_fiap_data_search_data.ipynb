```python
result_final_audio_features = []

generos = [
{"category_id": 65, "category": "Jungle"},
{"category_id": 66, "category": "Dub"},
{"category_id": 67, "category": "Roots Rock"},
{"category_id": 68, "category": "Southern Rock"},
{"category_id": 69, "category": "Jam Band"},
{"category_id": 70, "category": "Psychedelic Rock"},
{"category_id": 71, "category": "Prog Rock"},
{"category_id": 72, "category": "Classic"}
]


for idx, search in enumerate(generos):
    results = sp.search( search["category"] , limit=5 , offset=0 , type='playlist')
    print(search)

    #print(results)
    count_playlist = 0

    for idx, playlist in enumerate(results['playlists']['items']):
        count_playlist += 1
        #print(playlist['id'])
        playlist_items = sp.playlist_items( playlist['id'])    
        #print(playlist_items['items'][0]['track']['id'])

        for idx, item in enumerate(playlist_items['items']):
            if item['track'] is not None and item['track']['id'] is not None:
                print(item['track']['id'])
                audio_features = sp.audio_features(tracks=[item['track']['id']])
                if audio_features[0] is not None:
                    audio_features[0]["playlist_id"] = playlist['id']
                    #add categoria e data de criao na lista
                    audio_features[0]["category"] = search["category"]
                    audio_features[0]["category_id"] = search["category_id"]
                    audio_features[0]["creation_date"] = datetime.now()
                    
                    result_final_audio_features += audio_features 
        print(count_playlist)

            
jsonStr = json.dumps(result_final_audio_features, default=str)
#print(jsonStr)
df = pd.read_json(jsonStr)

#df_result = df.groupby('id').head(1)

df_result_drop_columns = df.drop(["type", "uri", "track_href", "analysis_url", "duration_ms", "time_signature"], axis= 1)

#print(df_result_drop_columns)

df_result_drop_columns.fillna(df_result_drop_columns.median(), inplace=True)

#df.drop_duplicates(subset = ['id', 'playlist_id'])
df_result_drop_columns.to_csv('excel_generos_65_72_.csv')

print(" ------------------------ Fim ------------------------------")


    
```
